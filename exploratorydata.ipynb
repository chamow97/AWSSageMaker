```python
import matplotlib.pyplot as plt
# Dataset
scores = [45, 50, 55, 60, 65, 70, 75, 80, 200]
# Create histogram
plt.hist(scores, bins=8, color='skyblue', edgecolor='black')
plt.title('Distribution of Exam Scores')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.show()

```

```python
import seaborn as sns
import pandas as pd
# Dataset
scores = pd.DataFrame({'Scores': [45, 50, 55, 60, 65, 70, 75, 80, 200]})
# Create box plot
sns.boxplot(x=scores['Scores'])
plt.title('Box Plot of Exam Scores')
plt.xlabel('Scores')
plt.show()

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
scores = np.array([45, 50, 55, 60, 65, 70, 75, 80, 200])

# Convert to pandas Series
scores_series = pd.Series(scores)

# Calculate rolling mean with a window of 3
rolling_mean = scores_series.rolling(window=3).mean()

# Plot original scores and rolling mean
plt.plot(scores, label='Original Scores', marker='o')
plt.plot(rolling_mean, label='Rolling Mean', linestyle='--', color='red')
plt.title('Exam Scores with Rolling Average')
plt.xlabel('Index')
plt.ylabel('Scores')
plt.legend()
plt.show()  

```
